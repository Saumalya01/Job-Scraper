```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
service_obj=Service(r"C:\Users\dell\Downloads\chromedriver-win32\chromedriver-win32\chromedriver.exe")
driver=webdriver.Chrome(service=service_obj)

from selenium.webdriver.common.by import By
import time
```

```python
driver.get('https://www.linkedin.com/jobs/search?location=India&geoId=102713980&f_JT=F%2CC&f_WT=2&currentJobId=3780010208&position=1&pageNum=0')
```

```python

import pandas as pd
import os

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
#Find number of job listings
y=driver.find_elements(By.CLASS_NAME,'results-context-header__job-count')[0].text
type(y)
import re

string = y

# Remove the "+" symbol
number_string = re.sub(r"\+$", "", string)

# Replace the comma with an empty string
number_string = number_string.replace(",", "")

# Try converting the string to an integer
try:
    number = int(number_string)
except ValueError:
    # Handle the case where the string cannot be converted to an integer
    print(f"Error: '{string}' cannot be parsed as an integer.")
    number = None

# Print the parsed integer
if number is not None:
    print(f"Parsed integer: {number}")
```

```python
#Loop to scroll through all jobs and click on see more jobs button for infinite scrolling

i = 2
while i <= int((number+200)/25)+1: 
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i = i + 1
    
    try:
        send=driver.find_element_by_xpath("//button[@aria-label='Load more results']")
        driver.execute_script("arguments[0].click();", send)   
        time.sleep(3)
  

            
    except:
        pass
        time.sleep(5)
         
```

```python
#Create empty lists for company name and job title

company_name= []
job_title= []
job_location= []
#job_description=[]
salary_range=[]
job_listing_date=[]

#Find company name and append it to the blank list

try:
    for i in range(number):
        company=driver.find_elements(By.CLASS_NAME,'base-search-card__subtitle')[i].text
        company_name.append(company)
except IndexError:
    print("no")
    
#Find job title name and append it to the blank list

try:
    for i in range(number):
        title=driver.find_elements(By.CLASS_NAME,'base-search-card__title')[i].text
        job_title.append(title)
except IndexError:
    print("no")
    
#Find job location and append it

try:
    for i in range(number):
        title=driver.find_elements(By.CLASS_NAME,'job-search-card__location')[i].text
        job_location.append(title)
except IndexError:
    print("no")
    
#Find salary range and append it

try:
    for i in range(number):
        title=driver.find_elements(By.CLASS_NAME,'job-search-card__salary-info')[i].text
        salary_range.append(title)
except IndexError:
    print("no")
    
#Find job listing date and append it

try:
    for i in range(number):
        title=driver.find_elements(By.CLASS_NAME,'job-search-card__listdate')[i].text
        job_listing_date.append(title)
except IndexError:
    print("no")
    
#Find job description and append it

#try:
    #for i in range(n):
        #title=driver.find_elements_by_class_name('')[i].text
        #job_description.append(title)
#except IndexError:
    #print("no")
    
#Find job links and append it to a list
linklist=[]
joblink = driver.find_elements(By.CLASS_NAME,'base-card__full-link')
for e in joblink:
    linklist.append(e.get_attribute('href'))
```

```python
#Create dataframe for company name and title

company=pd.DataFrame(company_name,columns=["company_name"])
title=pd.DataFrame(job_title,columns=["job_title"])
location=pd.DataFrame(job_location,columns=["job_location"])
sal=pd.DataFrame(salary_range,columns=["salary_range"])
date=pd.DataFrame(job_listing_date,columns=["job_listing_date"])
Job_url=pd.DataFrame(linklist,columns=["joblinks"])
x=company.join(title)
y=x.join(location)
z=y.join(sal)
a=z.join(date)

output=a.join(Job_url)

output.to_csv('linkedin_jobs.csv')
```
